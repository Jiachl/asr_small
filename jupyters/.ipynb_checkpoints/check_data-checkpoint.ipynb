# Test whisper model

```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from datasets import Audio, load_dataset
import evaluate
from transformers.models.whisper.english_normalizer import BasicTextNormalizer
import time
```

```python
processor = WhisperProcessor.from_pretrained("openai/whisper-medium")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-medium")
forced_decoder_ids = processor.get_decoder_prompt_ids(language="german", task="transcribe")
metric = evaluate.load("wer")
normalizer = BasicTextNormalizer()
ds = load_dataset("mozilla-foundation/common_voice_16_0", "de", split="test", streaming=True, token="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")
ds = ds.cast_column("audio", Audio(sampling_rate=16_000))
```

```python
wer, norm_wer = [], []
elapse = []
preds, labels = [], []
norm_preds, norm_labels = [], []
i = 0
init = time.time()
for data in iter(ds):
    start = time.time()
    input_speech = data["audio"]
    input_features = processor(input_speech["array"], sampling_rate=input_speech["sampling_rate"], return_tensors="pt").input_features
    predicted_ids = model.generate(input_features, forced_decoder_ids=forced_decoder_ids)
    cur_pred = processor.batch_decode(predicted_ids, skip_special_tokens=True)
    preds += cur_pred
    labels.append(data['sentence'])
    norm_preds += [normalizer(p).strip() for p in cur_pred]
    norm_labels += [normalizer(lb).strip() for lb in [data['sentence']]]

    end = time.time()
    elapse.append(f"{end - start:.2f}")
    
    if i % 1000 == 0:
        cur = time.time()
        wer = 100 * metric.compute(predictions=preds, references=labels)
        norm_wer = 100 * metric.compute(predictions=norm_preds, references=norm_labels)
        print(f"step {i}: wer {wer:.4f}, normalized wer {norm_wer:.4f} time elapse: {cur - init:.2f}")
        init = time.time()
    i += 1
        
wer = 100 * metric.compute(predictions=preds, references=labels)
norm_wer = 100 * metric.compute(predictions=norm_preds, references=norm_labels)
```
